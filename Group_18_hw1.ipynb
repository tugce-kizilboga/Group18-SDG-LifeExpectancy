```python
cols_to_plot = ['Life expectancy ', 'Schooling', 'Total expenditure', 'Adult Mortality', 'Status']
df_subset = df[cols_to_plot].dropna()

# Pairplot oluÅturma
pair_plot = sns.pairplot(df_subset, hue='Status', palette='husl', diag_kind='kde')
plt.show()
```

```python
numeric_cols = ['Life expectancy ', 'Schooling', 'Total expenditure', 'Adult Mortality', 'GDP']
corr_matrix = df[numeric_cols].corr()

# IsÄ± haritasÄ± oluÅturma
plt.figure(figsize=(10, 8))
# 'cmap' parametresini 'coolwarm' olarak deÄiÅtiriyoruz.
# 'vmin=-1, vmax=1' parametreleri renk skalasÄ±nÄ±n tam aralÄ±ÄÄ±nÄ± (-1 ile +1) kapsamasÄ±nÄ± saÄlar.
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)

# BaÅlÄ±k
plt.title('DeÄiÅkenler ArasÄ±ndaki Korelasyon IsÄ± HaritasÄ± (Coolwarm Paleti)', fontsize=14)

plt.show()
```

```python
plt.figure(figsize=(10, 8))

# 'annot=True' ile korelasyon katsayÄ±larÄ±nÄ± karelerin iÃ§ine yazdÄ±rÄ±yoruz.
# 'cmap' olarak 'YlGnBu' (SarÄ±-YeÅil-Mavi) paletini kullanarak pozitif iliÅkileri vurguluyoruz.
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=0.5)

plt.title('Korelasyon IsÄ± HaritasÄ±: SaÄlÄ±k ve EÄitim GÃ¶stergeleri (SDG 3 & 4)', fontsize=14)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 1. Grafik: YaÅam Beklentisi (SDG 3) YoÄunluk DaÄÄ±lÄ±mÄ±
sns.kdeplot(data=df, x='Life expectancy ', hue='Status', fill=True, ax=axes[0], palette='crest')
axes[0].set_title('YaÅam Beklentisi YoÄunluk DaÄÄ±lÄ±mÄ±', fontsize=12)

# 2. Grafik: OkullaÅma SÃ¼resi (SDG 4) YoÄunluk DaÄÄ±lÄ±mÄ±
sns.kdeplot(data=df, x='Schooling', hue='Status', fill=True, ax=axes[1], palette='flare')
axes[1].set_title('OkullaÅma SÃ¼resi YoÄunluk DaÄÄ±lÄ±mÄ±', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
df['Social_Investment_Index'] = df['Schooling'] * df['Total expenditure']

# Yeni sÃ¼tunla birlikte ilk 5 satÄ±ra bakalÄ±m:
print("Yeni OluÅturulan Sosyal YatÄ±rÄ±m Endeksi (Ä°lk 5 SatÄ±r):")
df[['Country', 'Year', 'Schooling', 'Total expenditure', 'Social_Investment_Index']].head()
```

```python
investment_comparison = df.groupby('Status')['Social_Investment_Index'].mean().sort_values(ascending=False)

print("StatÃ¼ye GÃ¶re Sosyal YatÄ±rÄ±m Endeksi OrtalamasÄ±:")
print(investment_comparison)

# GÃ¶rselleÅtirme
plt.figure(figsize=(8, 6))
sns.barplot(x=investment_comparison.index, y=investment_comparison.values, palette='plasma')

plt.title('GeliÅmiÅlik StatÃ¼sÃ¼ne GÃ¶re Sosyal YatÄ±rÄ±m GÃ¼cÃ¼', fontsize=14)
plt.ylabel('Ortalama Sosyal YatÄ±rÄ±m Endeksi (EÄitim * SaÄlÄ±k)', fontsize=12)
plt.xlabel('Ãlke StatÃ¼sÃ¼', fontsize=12)

plt.show()
```
